```python
import pandas as pd

INPUT_FILE = "jobs.csv"


def load_data(path):
    df = pd.read_csv(path)
    return df


def top_companies(df):
    return df["company"].value_counts().head(10)


def top_locations(df):
    return df["location"].value_counts().head(10)


def extract_skills(df):
    skills_series = df["tags"].dropna().str.split(", ")
    skills = skills_series.explode()

    return skills.value_counts().head(15)


def salary_summary(df):
    salary_df = df[["salary_min", "salary_max"]].dropna()

    if salary_df.empty:
        return None

    salary_df["avg_salary"] = (
        salary_df["salary_min"] + salary_df["salary_max"]
    ) / 2

    return salary_df["avg_salary"].describe()


def main():

    df = load_data(INPUT_FILE)

    print("\nTop Companies")
    print(top_companies(df))

    print("\nTop Locations")
    print(top_locations(df))

    print("\nTop Skills")
    print(extract_skills(df))

    print("\nSalary Summary")
    print(salary_summary(df))


if __name__ == "__main__":
    main()
```
