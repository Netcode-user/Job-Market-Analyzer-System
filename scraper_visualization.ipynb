```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Load dataset
df = pd.read_csv("output/jobs.csv")

# -------------------------------
# Data Cleaning
# -------------------------------
df['salary_min'] = pd.to_numeric(df['salary_min'], errors='coerce')
df['salary_max'] = pd.to_numeric(df['salary_max'], errors='coerce')

# Fill missing values for safe plotting
df['job_title'] = df['job_title'].fillna("Unknown")
df['location'] = df['location'].fillna("Unknown")
df['tags'] = df['tags'].fillna("")
```

```python

# -------------------------------
# 1. Bar Graph - Top Job Roles
# -------------------------------
plt.figure()
df['job_title'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Job Roles")
plt.xticks(rotation=45)
plt.show()

```

```python
# -------------------------------
# 2. Pie Chart - Top Locations
# -------------------------------
plt.figure()
df['location'].value_counts().head(5).plot(kind='pie', autopct='%1.1f%%')
plt.title("Top Locations")
plt.ylabel("")
plt.show()
```

```python

# -------------------------------
# 3. Line Chart - Jobs Distribution
# -------------------------------
plt.figure()
df['job_title'].value_counts().head(10).plot(kind='line', marker='o')
plt.title("Job Trends (Top Roles)")
plt.xticks(rotation=45)
plt.show()
```

```python

# -------------------------------
# 4. Histogram - Salary Distribution
# -------------------------------
plt.figure()
df['salary_min'].dropna().plot(kind='hist', bins=15)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.show()

```

```python
# -------------------------------
# 5. Box Plot - Salary Spread
# -------------------------------
plt.figure()
plt.boxplot(df['salary_min'].dropna())
plt.title("Salary Spread (Min Salary)")
plt.show()
```

```python

# -------------------------------
# 6. Scatter Plot - Salary Range
# -------------------------------
valid_df = df.dropna(subset=['salary_min', 'salary_max'])

plt.figure()
plt.scatter(valid_df['salary_min'], valid_df['salary_max'])
plt.xlabel("Min Salary")
plt.ylabel("Max Salary")
plt.title("Salary Scatter")
plt.show()
```

```python

# -------------------------------
# 7. Skills Analysis (Bar Graph)
# -------------------------------
all_tags = df['tags'].str.split(", ")
flat_tags = [tag for sublist in all_tags for tag in sublist if tag]

tag_counts = Counter(flat_tags)
top_tags = dict(tag_counts.most_common(10))

plt.figure()
plt.bar(top_tags.keys(), top_tags.values())
plt.title("Top Skills")
plt.xticks(rotation=45)
plt.show()

```

```python
# -------------------------------
# 8. Horizontal Bar - Companies Hiring
# -------------------------------
plt.figure()
df['company'].value_counts().head(10).plot(kind='barh')
plt.title("Top Hiring Companies")
plt.show()
```

```python
import matplotlib.pyplot as plt

top_jobs = df['job_title'].value_counts().head(5)

plt.figure()
top_jobs.plot(kind='area')

plt.title("Job Distribution Area Plot")
plt.xlabel("Job Title")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python

# -------------------------------
# 10. Correlation Heatmap (Manual - No seaborn)
# -------------------------------
corr = df[['salary_min', 'salary_max']].corr()

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.show()

print("✅ All visualizations displayed successfully!")
```
